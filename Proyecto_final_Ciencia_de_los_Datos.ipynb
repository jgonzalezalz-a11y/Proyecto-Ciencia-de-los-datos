```
df = df.drop(['student_id', 'salary_package_lpa', 'sleep_hours'], axis=1)
```

```
df.info()
```

```
df['volunteer_experience'] = df['volunteer_experience'].map({'Yes': 1, 'No': 0})
```

```
df['volunteer_experience'].value_counts()
```

```
df['gender'] = df['gender'].map({'Male': 1, 'Female': 0})
```

```
df['gender'].value_counts()
```

```
df['placement_status'] = df['placement_status'].map({'Placed': 1, 'Not Placed': 0})
```

```
df['placement_status'].value_counts()
```

```
df.info()
```

```
df['college_tier'] = df['college_tier'].map({'Tier 1': 1, 'Tier 2': 2, 'Tier 3': 3})
```

```
df['college_tier'].value_counts()
```

```
df = pd.get_dummies(df, columns=['branch'])
```

```
df.shape
```

```
df.iloc[:, -5:].head()
```

```
df = df.astype(int, errors='ignore')
```

```
df.iloc[:, -5:].head()
```

```
df.info()
```

```
plt.figure(figsize=(16, 10))
```

```
sns.heatmap(df.corr(), annot=False, cmap='RdYlGn', center=0)

plt.title('Mapa de Calor: ¿Qué variables están conectadas?')
plt.show()
```

```
ranking_exito = df.corr()['placement_status'].sort_values(ascending=False)
```

```
ranking_exito.head(10)
```

```
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.barplot(x='placement_status', y='internships_count', data=df, ax=axes[0], palette='Blues')
axes[0].set_title('Impacto de Pasantías')
sns.barplot(x='placement_status', y='projects_count', data=df, ax=axes[1], palette='Greens')
axes[1].set_title('Impacto de Proyectos')
sns.barplot(x='placement_status', y='coding_skill_score', data=df, ax=axes[2], palette='Oranges')
axes[2].set_title('Impacto de Coding Score')
plt.tight_layout()
plt.show()
```
